## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Research Question**

Which pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring based on their SEO and content performance signals?

**Unit of Analysis:**

An individual web page. Each page will be evaluated independently using its SEO and content performance signals to determine its priority for review.

**Decision**

Determine which pages should be prioritized for review.

**Action**

Create a ranked review queue that recommends whether a page should be refreshed, expanded, protected, pruned, or monitored.

**Cost of a Wrong Recommendation**

A false positive may cause unnecessary effort on pages that do not need attention, while a false negative may allow valuable pages to continue declining without intervention.

**Why Data or ML Can Help**

A page's priority depends on multiple signals, such as impressions, CTR, content age, ranking position, and update history. Machine learning can combine these signals to produce a more effective ranking than a simple rule-based approach.
