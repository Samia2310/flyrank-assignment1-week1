## 2. Target or proxy

The starter dataset does not contain a direct label indicating whether a page should be refreshed. Therefore, I will use trend_direction as a proxy target. Pages with the value "down" will be treated as declining pages that may require review.

```python
import os, sys, subprocess
import pandas as pd

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
else:
    while not os.path.isdir("data/raw") and os.getcwd() != "/":
        os.chdir("..")

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df["target"] = (df["trend_direction"].str.lower() == "down").astype(int)

df[["trend_direction", "target"]].head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why ML Beats a Fixed Rule**

A simple rule, such as selecting only stale pages with high impressions, cannot capture the combined effects of multiple SEO signals. Machine learning can learn relationships among impressions, CTR, ranking position, content age, and update history, producing a more effective ranking than a fixed rule. The starter notebooks demonstrated this by showing that the Random Forest model achieved approximately Precision@50 = 0.74, compared with 0.24 for the baseline rule.

```python
baseline_precision = 0.24
rf_precision = 0.74

print(f"Baseline Precision@50: {baseline_precision}")
print(f"Random Forest Precision@50: {rf_precision}")
print(f"Improvement: {rf_precision / baseline_precision:.1f}x")
```
